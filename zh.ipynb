```python
import requests as rq
from bs4 import BeautifulSoup as bs
from time import sleep
import json
import os
import re
from urllib import parse

listData = []
bookList = []
list1 = []
list2 = []
list3 = []
def home():
    url = 'https://www.bookwormzz.com'
    zh = '/zh/'
    response = rq.get(url+zh)
    result = parse.unquote(response.text)
#     print(result)
    soup = bs(result,"lxml")
    
    for a in soup.select('a'):
        if a.has_attr('data-ajax'):
#             print(a['href'])
#             list1.append(a['href'])
            bookName = a.text
#             print(bookName)
            bookList.append(bookName)
            reg01 = r'(/(.)+\.php\/0\.xhtml)'
            match01 = re.search(reg01,a['href'])
#             print(match01[0])
            list1.append(match01[1])

#     print(list1)
#     print(bookList)
    for book in list1:
        path01 = f"{url}{book}#book_toc"
#         print(url+book+'#book_toc')
#         print(path01)
        list2.append(path01)
#     print(list2)

    for a in list2:
        bResponse = rq.get(a)
#         print(bResponse.text)
#         print("="*50)
        bsoup =bs(bResponse.text,"lxml")
        for ch in bsoup.select('a'):
            if ch.has_attr('data-icon'):
                continue
            else:
                list3.append(ch['href'])
#                 print(ch['href'])
        
#     print(list3)
        for chap in list3:  # list3 迴圈可以放在list2 else裡面就好
            cpath = f"{url}{chap}"
            cres = rq.get(cpath)
            csoup = bs(cres.text,"lxml")
        
            cContent = csoup.select("div#html")[0].get_text()
            ctitle = csoup.select("h3")[0].text
            cc = cContent.replace("　　","")
            ccc= cc.replace("\n","")
            
            cregx= r'/(.)+\/'
            cmatch = re.search(cregx,chap)
            cind = cmatch[0].replace("/","")
            cindex = cind.replace(".php","")
#         print()
#         print("="*60)
#         print(cc)
#         print("="*60)
#         print()
            fj= open(f"C:/Users/User/python_web_scraping-master/zhjson/{cindex}_{ctitle}.json","w",encoding="utf-8")
            fj.write(ccc)
            fj.close()
            fp= open(f"C:/Users/User/python_web_scraping-master/zhtxt/{cindex}_{ctitle}.txt","w",encoding="utf-8")
            fp.write(ccc)
            fp.close()
if __name__ == '__main__':
    home()
```
